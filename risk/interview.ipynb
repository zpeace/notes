
##### How did you built Draw and Curtailment model?

We tried to model the draw and curtailment behavior of our customers. It is an redesign of our previous module. We had probability models for draw, curtailment and no action. Lookup table were used to decide the amount. A spearate draw spike table to predict draw amount when transition from 1DPD to 30DPD.

1. Drawbacks of previous model. 
    - no macro econ sensitivity in amount module
    - deteriation of performance on new sample
    - definition of draw/curtailment incorrect
    
2. Data creation.
    - use cashflow to define draw and curtailment with payment type considered
    - findings of inconsistency: draw term missing; open indicator change from zero to one.
    - creation of draw curtailment related variables. lag draw, lag draw count, amount, open to buy
    
3. Exploratory data anaysis
    - draw/curtail probability on different risk dimensions
    - univariate analysis
    - bivariate analysis
    - clustering analysis
    - decision tree
    
4. Model design
    - draw spike table not applicable. We may think that customer who goes to 30dpd might draw a large amount in expection of blocking. 
    - previously 2 probability models for draw and 2 probability models for curtailment
    - tried even more model (rookie move), 7 models for probability segement on 0 balance, 0 otb, from delq state. Too complicated. While spbt looks OK, mpbt showed bad performance.
    - back 1 model for draw probability, 1 for curtail.
    - linear models for draw percentage and curtailment percentage. 
    
5. Model fitting
    - stepwise 
    - tunning bins
    - discover interaction (plot on risk dimentions by H1 H2)
    - monotonicity and explanability
    - vifs checked for mulicolinearity
    
6. Performance analysis
    - SPBT by time and main risk dimensions (product, heritage, cohort, FICO, LTV, HPA)
    - MPBT: error propagation
    
7. Implementation testing
    - step by step (in time) comparison of cash flow and variable logic
    - customized shell script to compare multiple versions of different products, cohorts conveniently
    
8. Key risk drivers
    - FICO, HERITAGE, product, utiliztion, property state, HPA, OTB, UPB
